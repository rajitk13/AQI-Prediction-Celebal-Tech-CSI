```r
library(tidyverse)
```

```r
data<-read.csv('/content/sample_data/one_month_dehradun.csv')
print(data)
```

```r
str(data)
```

```r
glimpse(data)
```

```r
ggplot(data, aes(x = aqi, y = pm25 , color = aqi)) +
  geom_point()
```

```r
ggplot(data, aes(aqi, ts)) +
  geom_point() +
  facet_wrap(~aqi)
```

```r
ggplot(data, aes(aqi, pm25)) +
  geom_point() +
  geom_smooth()
```

```r
ggplot(data, aes(aqi, ts)) +
  geom_point() +
  geom_smooth(span = 0.2)


ggplot(data, aes(aqi, ts)) +
  geom_point() +
  geom_smooth(span = 1)
```

```r
ggplot(data, aes(aqi, pm25)) + geom_jitter()
ggplot(data, aes(aqi, ts)) + geom_boxplot()
ggplot(data, aes(aqi, no2)) + geom_violin()
```

```r
ggplot(data, aes(aqi)) +
  geom_freqpoly(binwidth = 2.5)
ggplot(data, aes(pm25)) +
  geom_freqpoly(binwidth = 1)
```

```r
ggplot(data, aes(aqi, colour = ts)) +
  geom_freqpoly(binwidth = 0.5)
ggplot(data, aes(aqi, fill = ts)) +
  geom_histogram(binwidth = 0.5) +
  facet_wrap(~ts, ncol = 1)
```

```r
ggplot(data, aes(aqi, pm25)) +
  geom_path() +
  geom_point()
```
