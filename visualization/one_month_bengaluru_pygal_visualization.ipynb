```python
import pygal
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/celebal internship/air quality index/bengaluru_aqi/one_month_bengaluru.csv')
```

```python
pygal.Bar().add('AQI', df['aqi'][:10])
```

```python
pygal.Bar().add('PM 2.5', df['pm25'][:10])
```

```python
line_chart = pygal.Line()
line_chart.add('aqi', df['aqi'])
```

```python
funnel_chart = pygal.Funnel()
funnel_chart.title = 'AQI'
funnel_chart.add('AQI', df['aqi'][:15])
funnel_chart.add('PM 2.5', df['pm25'][:15])
```
